# Brief introduction to Machine Learning

Machine learning emerged from inquiries in statistics, computer science, information theory, artificial intelligence, and pattern recognition. We can think of it as sets of tools for investigating, modeling, and understanding data. What are the steps? 

1. Formulate a question. Do you want to predict (or infer) something?  

2. What is your task?
    - Supervised or unsupervised?
    - Classification or regression?  

3. Find some data  

4. Identify appropriate algorithms  

5. Handle missing data  

6. Split the data  

7. Fit the model to the training set and evaluate training performance  

8. Fit the model to the test set and evaluate test performance  

### 1. Formulate a question. Do you want to predict (or infer) something?  

**Supervised Learning:** Machine learning is great for predicting an outcome (dependent variable) based on some covariates (input variables). An example could include predicting whether or not someone has diabetes or not based on other health variables such as sex, age, cholesterol levels, body mass, smoker, etc. to predict this health outcome. 

**Unsupervised Learning:** Machine learning is also great for finding patterns in data without trying to predict or infer anything. Instead, we can just see how the data sort on their own accord! We could use consumer data to see if there if groups of shoppers emerge based on patterns in income, spending, sales, and time of year. 

### 2. What is your task?

What are your goals? What is your question? What do you want to learn? 

##### Supervised or unsupervised?

**Supervised syntax** looks like this: Y ~ X

Y = dependent/target/outcome/response variable  
X = independent/predictor/input variable  

In supervised machine learning, the training dataset is used to train a model that predicts the outputs of a target function, or an estimation of the actual but unknown function that maps X to Y. Here, the model learns the characteristics of the data on a training dataset and its performance is evaluated using a "performance metric" - a metric that describes how well it could predict the outcomes (accuracy, MSE, RMSE, AUC, etc.). This gives us a way to measure the goodness of fit for the model on the training data.  

However, we also want to see how well it performs on the test dataset, or data it has not yet seen! The model performs well if it has similar performance on the test dataset it has never seen before, because it can be generalized to new data - this is a major strength of machine learning.  

If the model performs poorly on the training dataset it is said to be **underfit** because it was not able to learn about relationships between the X and Y variables. If it performs well on the training set but poorly on the test set the model is said to be **overfit** because the model performed worse than expected when given new data (although patterns might be due to noise).  

This can be extended to unsupervised learning as well, where there is no Y variable - the target function seeks to identify patterns in the data rather than predicting an outcome. 

**Unsupervised syntax** looks like: **~ X**

##### Classification or regression? 

**Classification** is used when the Y outcome variable is categorical. "yes" or "no" is a _binary_ example: 1 is prediction of the "yes" category and 0 as the "no". This can be extended to multi-level classification as well.  

**Regression** is used when the Y outcome variable is continuous and we want the model to predict it using X variable(s).

### 6. Split the data.  
[Split the data](https://en.wikipedia.org/wiki/Training,_validation,_and_test_sets) into training and test sets. [Data splitting](https://www.mff.cuni.cz/veda/konference/wds/proc/pdf10/WDS10_105_i1_Reitermanova.pdf) is a fundamental preprocessing step used to divide a dataset into "training" and "test" sets. The majority portion of the data (say, 75%) is assigned to the training set, while the remaining 25% of data is assigned to the test set.  

### 7. Fit the data to the training set.  

Recall that we want to train the model on the "training set" to see how well it learns the relationships between X and Y by evaluating its performance. 

##### Evaluate performance. 
Evaluate performance. [k-fold cross validation](https://en.wikipedia.org/wiki/Cross-validation_(statistics)) is preferred because it repeats this splitting process **k** number of times, where each fold  serves as the training set **k-1** times and as the test set exactly once. 

Thus, in 4-fold cross validation, the data are divided into four equal sized chunks. In the first iteration, the first 25% of the data is the test set, while the remaining 75% is the training set. In the second iteration, the second 25% is the test set and the other 75% is the training set, and so on:

Other performance metrics include [sensitivity and specificity](https://en.wikipedia.org/wiki/Sensitivity_and_specificity)

# The value of understanding simple linear regression - by hand! 

Doing a simple [OLS](https://en.wikipedia.org/wiki/Ordinary_least_squares) regression by hand provides a way to understand what supervised machine learning is doing under the hood. OLS regression builds a model that tries to predict Y using X (regresses Y onto X). 

First, let's generate toy predictor (X) and response (Y) variables and compute their means. This will be our "training set":

```python
# Generate toy predictor (X) and response (Y) variables
X = np.array([2, 4, 8, 12, 18, 20])
Y = np.array([1, 3, 5, 9, 19, 21])
print(X)
print(Y)

## Calculate their means
mean_X = round(sum(X) / len(X), 2)
mean_Y = round(sum(Y) / len(Y), 2)
print("mean of X is:", mean_X)
print("mean of Y is:", mean_Y)
```

### Vectorized operations

We can also perform "vectorized" operations, such as subtracting the mean of X from each X value or the mean of Y from each Y value. That is, we can do math on arrays of numbers simultaneously: 

```python
print(X-mean_X)
print(Y-mean_Y)
```

This is important bcause it helps us calculate our [beta coefficeints](https://en.wikipedia.org/wiki/Standardized_coefficient) in the regression. 

```python
## Estimate the B1 coefficient (slope)
B1 = sum((X-mean_X) * (Y-mean_Y)) / sum((X-mean_X)**2)
print("slope is equal to", round(B1,2))

## Estimate B0 coefficient (intercept)
B0 = mean_Y - (B1 * mean_X)
print("intercept is equal to", round(B0, 2))
```

# Challenge 2

1. Use this website to (https://seaborn.pydata.org/tutorial/regression.html) to learn how to make a scatterplot of X and Y (X on the x-axis and Y on the Y-axis) with a best fit line. 

```python
## YOUR CODE HERE
sns.lmplot(x="X", y="Y", data=df);
```

Now that we have calculated the OLS "best fit line" (the line that minimizes the sum of the squared errors) so that we can generate predicted values (our "test set") and assess the performance of the model:

```python
## Generate predicted Y values by plugging our X values into the equation: 
Y_hat = B0 + B1 * X
print(Y_hat)
print(Y)
```

Our performance metric will be [root mean square error (RMSE)](https://en.wikipedia.org/wiki/Root-mean-square_deviation), or the standard deviation of the residuals (aka prediction errors). Error is measured as the vertical distance from a data point to the best fit line. 

However, we have to do some calculations first before we get to RMSE! 

```python
## First, calculate the error for each observation by subracting the predicted value from it:
Y_err = Y - Y_hat
print(Y_err)
```

```python
## Second, calculate the square of each of these errors:
Y_err_sq = Y_err**2
print(Y_err_sq)
```

```python
## Third, sum these values
sum_squared_err = sum(Y_err_sq)
print(sum_squared_err)
```

```python
## Fourth, calculate the RMSE - take the square root of the summed squared error divided by the length of Y:
import math
RMSE = math.sqrt(sum_squared_err / len(Y))
print(round(RMSE, 5))
```

# CountVectorizer

`CountVectorizer` will help us quickly tokenize text, learn its vocabulary, and encode the text as a vector for use in machine learning. This is often referred to as document encoding.

# Document encoding / machine learning (continued)

[Term frequencyinverse document frequency (TFIDF)](https://en.wikipedia.org/wiki/Tf%E2%80%93idf) can be thought of as an extension of `CountVectorizer`. However, instead of counting words, TFIDF identifies unique words in text and across documents. 

# Vocabulary
- **Document Term Matrix: ** Is a matrix that describes the frequency of terms that occur in a collection of documents. In a document-term matrix, rows correspond to documents in the collection and columns correspond to terms.

- **TF-IDF Scores: ** Short for term frequencyinverse document frequency, is a numerical statistic that is intended to reflect how important a word is to a document in a collection or corpus.

- **Topic Modeling: ** A general class of statistical models that uncover abstract topics within a text. It uses the co-occurrence of words within documents, compared to their distribution across documents, to uncover these abstract themes. The output is a list of weighted words, which indicate the subject of each topic, and a weight distribution across topics for each document.
    
- **LDA: **Latent Dirichlet Allocation. A particular model for topic modeling. It does not take document order into account, unlike other topic modeling algorithms.

# TF-IDF scores

How to find distinctive words in a corpus is a long-standing question in text analysis. Today, we'll learn one simple approach to this: TF-IDF. The idea behind words scores is to weight words not just by their frequency, but by their frequency in one document compared to their distribution across all documents. Words that are frequent, but are also used in every single document, will not be distinguishing. We want to identify words that are unevenly distributed across the corpus using TFIDF. By offsetting the frequency of a word by its document frequency (the number of documents in which it appears) will in theory filter out common terms such as 'the', 'of', and 'and'.

Traditionally, the inverse document frequency is calculated as such:

**number_of_documents / number_documents_with_term**

so:

**tfidf_word1 = word1_frequency_document1 * (number_of_documents / number_document_with_word1)**

You can, and often should, normalize the numerator: 

**tfidf_word1 = (word1_frequency_document1 / word_count_document1) * (number_of_documents / number_document_with_word1)**

We can calculate this manually, but scikit-learn has a built-in function to do so. This function also uses log frequencies, so the numbers will not correspond excactly to the calculations above. We'll use the [scikit-learn calculation](http://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html). 

# TF-IDFVectorizer Function

To do so, we simply do the same thing we did before with `CountVectorizer`, but instead we use the function `TfidfVectorizer`

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidfvec = TfidfVectorizer()
sparse_tfidf = tfidfvec.fit_transform(reviews['body_without_digits'])
print(sparse_tfidf)
```

Turn this into a Pandas DataFrame:

# Fit the LDA model

Now we're ready to fit the model. This requires the use of `CountVecorizer`, which we've already used, as well as the `LatentDirichletAllocation` function.

See [here](http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.LatentDirichletAllocation.html) for more information about this function. 

First, we have to import it from sklearn. We tell the model how many topics we expect to find.

```python
from sklearn.decomposition import LatentDirichletAllocation
n_topics = 5
```

The input to LDA should be a DTM (with either counts or TF-IDF scores).

```python
# TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_df=0.80, min_df=50,
                                   max_features=5000,
                                   stop_words='english')
tfidf = tfidf_vectorizer.fit_transform(df_lit['text'])
```

[Check out this question](https://stackoverflow.com/questions/27697766/understanding-min-df-and-max-df-in-scikit-countvectorizer) to learn more about the `max_df` and `min_df` arguments. 

```python
# CountVectorizer
tf_vectorizer = CountVectorizer(max_df=0.80, min_df=50,
                                max_features=5000,
                                stop_words='english'
                                )
tf = tf_vectorizer.fit_transform(df_lit['text'])
```

# Finally! Fit the model

```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
lda = LatentDirichletAllocation(n_components=n_topics, max_iter=20, random_state=0)
lda = lda.fit(tf)
```

Below is a function to print out the top words for each topic in a pretty way. 

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("\nTopic #{}:".format(topic_idx))
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
print()
```

```python
tf_feature_names = tf_vectorizer.get_feature_names()
print_top_words(lda, tf_feature_names, 50)
```

# Intuiting the Logistic Regression Model

But what happens if one of your variable is categorical and not continuous? Suppose we don't care about the `Grade` score, but we just care if you `Pass` or not?

How would we fit a line to that? That's where the [logistic function](https://en.wikipedia.org/wiki/Logistic_function) can be handy. The general logistic function is:

$ f(x) = \frac{1}{1 + e^{-x}} $

# The `LogisticRegression` function

Let's go back to `sklearn`.

Let's save our logistic model in a variable named `lr`:

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver = "liblinear")
```

We'll `reshape` our arrays since scikit-learn prefers them in a specified format.

```python
X = np.array(study['Study_Hours']).reshape(-1,1)
Y = np.array(study['Pass']).reshape(len(study['Pass']),)
X, Y
```

We can use the `fit` function again on our `X` and `y`:

```python
lr.fit(X, Y)
```

Plot!

```python
sns.regplot(x = study.Study_Hours, y = study.Pass, logistic = True);
```
